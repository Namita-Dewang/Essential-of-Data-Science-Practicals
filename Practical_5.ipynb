```python
import pandas as pd
import matplotlib.pyplot as plt
d=pd.read_csv('drive/My Drive/Colab Notebooks/tips.csv')
print(d)
```

```python
t1 = d.groupby("sex").count()
t1["smoker"].plot(kind="pie",autopct = '%1.1f%%')
plt.show()
print(t1)
```

```python
t1 = d.groupby("size").count()
t1["smoker"].plot(kind = "pie",autopct = '%1.1f%%')
plt.show()
print(t1)
```

```python
t1 = d.groupby("day").count()
t1["smoker"].plot(kind = "pie",autopct = '%1.1f%%')
plt.show()
print(t1)
```

```python
t1 = d.groupby("tip").count()
t1["sex"].plot(kind = "pie",autopct = '%1.1f%%')
plt.show()
print(t1)
```
